```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values, avoiding x = 2 to prevent division by zero
x = np.linspace(-5, 5, 1000)
x = x[x != 2]  # remove the point where the function is undefined

# Define the function
y = (x**3 - 8) / (x - 2)

# Plot the function
plt.plot(x, y, label='(x³ - 8)/(x - 2)')

# Add a hole at x = 2 (optional for accuracy)
plt.plot(2, 12, 'ro', label='hole at x=2')  # since (2² + 2*2 + 4) = 12

# Add labels and grid
plt.title('Graph of (x³ - 8)/(x - 2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axvline(x=2, color='gray', linestyle='--', linewidth=0.5)  # optional: show the vertical line
plt.legend()
plt.show()
```
