```python
import numpy as np

#def f(x): return x**2 + (2*x)
def f(x): return 2*x

x = np.linspace(-5, 5, 100)

h = 0.0001

#cal_derivative = (f(x+h) - f(x-h))/(2*h)
der = (f(x) - f(x-h)) / h

der
```

```python
import matplotlib.pyplot as plt 

plt.plot(x, f(x), label='f(x)')
plt.plot(x, der, label="f'(x)")
plt.legend()
plt.show()
```
